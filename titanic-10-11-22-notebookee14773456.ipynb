## Removing Rows with null values

```python
train_data2 = train_data.dropna()
```

```python
train_data2.info()
```

```python
test_data1 = test_data.dropna()
test_data1.info()
```

## Encoding object type column into int64

```python
# not using onehotencoder as we have many data 

# from sklearn.preprocessing import OneHotEncoder
# enc = OneHotEncoder()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
# converting object type data into int64 type

for i in train_data2.columns:
    if train_data2[i].dtype == object :
#         print(i)
        train_data2[i] = le.fit_transform(train_data2[i])
#         print(train_data2[i])
```

```python
train_data2.info()
```

```python
for i in test_data1.columns:
    if test_data1[i].dtype == object:
#         print(i)
        test_data1[i] = le.fit_transform(test_data1[i])
```

```python
test_data1.info()
```

## splitting the data

```python
# primary splitting 

X = train_data2.drop('Survived', axis=1)
y = train_data2['Survived']
```

```python
X.info()
```

```python
# X.columns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

## Performanc of different model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
y_pred = knn.predict(X_val)
# y_pred.shape
print(classification_report(y_val, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0)
```

```python
rf.fit(X_train, y_train)
y_pred = rf.predict(X_val)
print(classification_report(y_val, y_pred))
```

```python
rf.predict(test_data1)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(random_state=0)
```

```python
dt.fit(X_train, y_train)
y_pred = dt.predict(X_val)
print(classification_report(y_val, y_pred))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier(random_state=0)
```

```python
gbc.fit(X_train, y_train)
y_pred = gbc.predict(X_val)
print(classification_report(y_val, y_pred))
```
